Preprocessing Data

```python
import nltk
import string
import re
```

```python
def case_folding(abstrak):
    abstrak = abstrak.lower()
    return abstrak
df['abstrak_lower'] = df['Abstrak'].apply(lambda x: case_folding(x))
```

```python
df.head()
```

```python
#import stopword
from nltk.corpus import stopwords 
nltk.download('stopwords')
stopwords_indonesia = stopwords.words('indonesian')
```

```python
#install and import pysastrawi
!pip install PySastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def cleaning(abstrak):
    symbols = "0123456789,!\"#$%&()*+-.…/:;<=>?@[\]^_`{|}~\n"
    abstrak = re.sub(r"\s—\s", "", abstrak)
    abstrak = re.sub(r"http\S+", "", abstrak)
    for i in symbols:
        abstrak = np.char.replace(abstrak, i, ' ')
    return abstrak
df['abstrak_clean'] = df['abstrak_lower'].apply(lambda x: cleaning(x))
```

```python
df.head()
```

```python
def split(abstrak):
    return abstrak.split()
df['abstrak_split'] = df['abstrak_clean'].apply(lambda x: split(x))
```

```python
df.head()
```

```python
def stemmingNormalize(comment):
    data = []
    for term in comment:
        term = stemmer.stem(term)
        data.append(term)
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
df['abstrak_stem'] = df['abstrak_split'].apply(stemmingNormalize)
```

```python
df.head()
```

```python
abstrak = df['Abstrak']
X = df['abstrak_stem']
y = df.values[:,0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(X)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components =10)

X = pca.fit_transform(X)

pca.explained_variance_ratio_
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
center = kmeans.cluster_centers_
center
```

```python
y_pred = kmeans.labels_

data = kmeans.labels_
res = []
for i in data:
    if i == 0:
        res.append("RPL")
    else:
        res.append("KOMPUTASI")
print(res)
```

```python
y_true = df['Kategori'].values.tolist()
print(df['Kategori'].values.tolist())
```

```python
n = len(y_true)
for i in range(n):
    print(res[i],"\t : ",y_true[i])
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_true, res))
```
